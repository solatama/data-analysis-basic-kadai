```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=2, ncols=3, tight_layout=True)

y0 = [1, 2, -5, 2]
ax[0, 0].plot(y0)

x0 = np.linspace(0, 10, 100)
y1 = 2 + 2*np.sin(2*x0)
ax[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))
ax[0, 1].plot(x0, y1)

x1 = np.random.normal(15, 5, 2000)
ax[0,2].hist(x1, bins=30)

np.random.seed(3)
x2 = 4 + np.random.normal(0, 2, 60)
y2 = 4 + np.random.normal(0, 2, len(x2))
ax[1, 0].scatter(x2, y2)
plt.show()
```
